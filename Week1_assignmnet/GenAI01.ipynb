```python
""" Required four-line import statement """
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

‚úîÔ∏è **Á®ãÂºèÁ¢ºË™™Êòé**

1. Áî¢Áîü0Âà∞2$œÄ$‰πãÈñìÁöÑ1000ÂÄãÁ≠âÈñìË∑ùÈªûÔºå‰ΩúÁÇ∫ÂèÉÊï∏t
2. ÂÆöÁæ©ÊÑõÂøÉÁöÑÂèÉÊï∏ÊñπÁ®ãÂºè: <br>
$x = 16 \sin^3(t)$ <br>
$y = 13 \cos(t) - 5 \cos(2t) - 2 \cos(3t) - \cos(4t)$




```python
""" ‰ΩøÁî®ÂèÉÊï∏ÊñπÁ®ãÂºèÁπ™Ë£ΩÊÑõÂøÉÂúñÂΩ¢ """

t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

plt.plot(x, y, 'r', linewidth=4)   # ‰ª•Á¥ÖÁ≠ÜÁï´ÂúñÔºåÁ∑öÂØ¨Ë®≠ÂÆöÁÇ∫4
plt.axis('off')             # Èö±ËóèÂùêÊ®ôËª∏
plt.title("Heart Shape", fontsize=12) # Ë®≠ÂÆöÂúñÁâáÊ®ôÈ°åÂèäÂ≠óÈ´îÂ§ßÂ∞è
```
